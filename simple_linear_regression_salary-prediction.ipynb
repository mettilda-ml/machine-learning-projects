```python
#Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("Salary_Data.csv") # from pandas libraries, it used to read the file from the path
```

```python
data # call the varible , to print
```

```python
data.head()
```

```python
data.head(10)
```

```python
independent = data[["YearsExperience"]] # Independent variable, years of Experiencce, datatypes
dependent = data[["Salary"]] # Dependent Variable, Salary, datatype float
```

```python
plt.scatter(independent,dependent)
plt.xlabel("YearsExperience",fontsize=20)
plt.ylabel("Salary",fontsize=20)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent,dependent, test_size = 1/3, random_state =0) # x -> input, y-> output
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print(type(x_train), x_train.shape)
print(type(y_train), y_train.shape)
```

```python
# Viewing the b1 and bo  value
weight=regressor.coef_
print("Weight of the Model={}".format(weight))
bais=regressor.intercept_
print("Intercept of the Model={}".format(bais))
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
